```python
import matplotlib.pyplot as plt

# Read the total rewards data from the file
with open('total_rewards.txt', 'r') as file:
    total_rewards = [float(line.strip()) for line in file]

# Calculate the cumulative reward
cumulative_rewards = [sum(total_rewards[:i+1]) for i in range(len(total_rewards))]

# Plot the learning curve
plt.plot(cumulative_rewards)
plt.xlabel('Time Step')
plt.ylabel('Cumulative Reward')
plt.title('Learning Curve')
plt.show()
```

```python
import numpy as np
import random

# Environment parameters
num_states = 3
num_actions = 3

# Q-learning parameters
alpha = 0.5
gamma = 0.9
epsilon = 0.1
num_episodes = 5000
max_steps_per_episode = 100

# Initialize the Q-table
Q_table = np.zeros((num_states, num_actions))

# Define a function to choose an action using an epsilon-greedy strategy
def choose_action(state, Q_table, epsilon):
    if random.uniform(0, 1) < epsilon:
        return random.choice(range(num_actions))
    else:
        return np.argmax(Q_table[state])

# Define a function to simulate the interaction between NodeA and NodeB
def environment(state, action):
    transmission_powers = [0, 1, 2]
    packet_loss_probabilities = [0.3, 0.1, 0.05]
    energy_consumptions = [1, 2, 4]

    transmission_power = transmission_powers[action]
    packet_loss_probability = packet_loss_probabilities[transmission_power]
    energy_consumption = energy_consumptions[transmission_power]

    successful_transmission = random.uniform(0, 1) > packet_loss_probability
    reward = (10 if successful_transmission else -10) - energy_consumption
    next_state = (state + action) % 3

    return next_state, reward

# Training progress tracking
episode_rewards = []
print_every = 100

# Train the Q-learning algorithm
for episode in range(num_episodes):
    state = random.choice(range(num_states))
    total_reward = 0

    for step in range(max_steps_per_episode):
        action = choose_action(state, Q_table, epsilon)
        next_state, reward = environment(state, action)
        total_reward += reward

        # Update the Q-table
        target = reward + gamma * np.max(Q_table[next_state])
        Q_table[state, action] = (1 - alpha) * Q_table[state, action] + alpha * target

        state = next_state

    episode_rewards.append(total_reward)

    if (episode + 1) % print_every == 0:
        avg_reward = sum(episode_rewards[-print_every:]) / print_every
        print(f"Episode: {episode + 1}, Average Reward: {avg_reward}")

# Save the trained Q-table to a file
np.savetxt("Q_table.txt", Q_table)
```
